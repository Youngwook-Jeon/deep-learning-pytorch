```python
import numpy as np
import torch
```

```python
# Using numpy
nv = np.array([[1, 2, 3, 4]])
print(nv)

print(nv.T)

nvT = nv.T
print(nvT.T)
```

```python
# Using pytorch
pt = torch.tensor([[1, 2, 3, 4]])
print(pt)

print(pt.T)

ptT = pt.T
print(ptT.T)
```

```python
# Examine dt
print(f'Variable nv is of type {type(nv)}')
print(f'Variable pt is of type {type(pt)}')
```
