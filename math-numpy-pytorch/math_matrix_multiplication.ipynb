```python
import numpy as np
import torch
```

```python
# Using numpy
A = np.random.randn(3, 4)
B = np.random.randn(4, 5)
C = np.random.randn(3, 7)

print(np.round(A@B, 2)), print(' ')
print(np.round(C.T@A, 2)), print(' ')
```

```python
# Using pytorch
A = torch.randn(3, 4)
B = torch.randn(4, 5)
C = torch.randn(3, 7)
D1 = np.random.randn(4, 5)
D2 = torch.tensor(D1, dtype=torch.float)

print(np.round(A@B, 2)), print(' ')
print(np.round(C.T@A, 2)), print(' ')
print(np.round(A@D1, 2)), print(' ')
print(np.round(A@D2, 2)), print(' ')
```
