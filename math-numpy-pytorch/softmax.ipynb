```python
import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
```

```python
z = [1, 2, 3]
num = np.exp(z)
den = np.sum(np.exp(z))
sigma = num / den

print(sigma)
print(np.sum(sigma))
```

```python
z = np.random.randint(-5, high=15, size=25)
print(z)

num = np.exp(z)
den = np.sum(num)
sigma = num / den

plt.plot(z, sigma, 'bo')
plt.xlabel('Original number (z)')
plt.ylabel('Softmaxified $\sigma$')
# plt.yscale('log')
plt.title('$\sum\sigma$ = %g'%np.sum(sigma))
plt.show()
```

# Using pytorch

```python
softfunc = nn.Softmax(dim=0)
sigmaT = softfunc(torch.Tensor(z))

print(sigmaT)
```

```python
# show that they are the same
plt.plot(sigma, sigmaT, 'bo')
plt.xlabel('"Manual" softmax')
plt.ylabel('Pytorch nn.Softmax')
plt.title(f'The two methods correlate at r={np.corrcoef(sigma,sigmaT)[0,1]}')
plt.show()
```
